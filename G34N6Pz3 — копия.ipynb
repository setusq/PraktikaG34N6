```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:\\Users\\Администратор\\Downloads\\titanic.csv')
```

```python
df
```

```python
X= df['Sex']
Y= df['Survived']
```

```python
survived_passengers = df[df['Survived'] == 1]

male_survived = survived_passengers[survived_passengers['Sex'] == 'male']
female_survived = survived_passengers[survived_passengers['Sex'] == 'female']

plt.figure(figsize=(10, 6))
plt.hist(male_survived['Age'], bins=20, alpha=0.5, label='Мужчины', color='blue')
plt.hist(female_survived['Age'], bins=20, alpha=0.5, label='Женщины', color='pink')
plt.xlabel('Возраст')
plt.ylabel('Количество выживших')
plt.legend()
plt.title('Зависимость выживших от возраста и пола')
plt.show()
```

```python
survival_by_class = df.groupby('PClass')['Survived'].sum()
plt.figure(figsize=(8, 6))
survival_by_class.plot(kind='bar', color='skyblue')
plt.xlabel('Класс пассажиров')
plt.ylabel('Количество выживших')
plt.title('Зависимость выживших от класса пассажиров')
plt.xticks(rotation=0)
plt.show()
```
